```python
import pandas as pd
from evidently import Report
from evidently.presets import DataDriftPreset

# 1. Load dataset
data_url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
boston_frame = pd.read_csv(data_url)

# 2. Split data
reference_data = boston_frame[:250]
current_data = boston_frame[250:]

# 3. Define the Report Template
report = Report(metrics=[DataDriftPreset()])

# 4. RUN the report and CAPTURE the result
# In 0.7+, .run() returns the object containing the calculated metrics
result = report.run(reference_data=reference_data, current_data=current_data)

# 5. Save the RESULT to HTML
# We call save_html on 'result', not 'report'
result.save_html("report.html")

print("Report successfully saved to report.html using Evidently 0.7+ API")
```
